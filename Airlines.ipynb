```python
# Import libraries 
import pandas as pd
import numpy as np
import altair as alt
```

```python
# Assign spreadsheet filename to `file`
file = 'statistic_id186821_sustainable-energy-investment_-wind-energy-technologies-2004-2017.xlsx'

# Load spreadsheet
xl = pd.ExcelFile(file)

# Load a sheet into a DataFrame by name: df1
df1 = xl.parse('Data')
df1
```

```python
import altair as alt
import pandas as pd

alt.Chart(df1).mark_bar().encode(
    x='Year',
    y='Investments:Q'
).interactive()
```

```python
alt.Chart(df1).mark_line().encode(
    x='Year:N',
    y='Investments:Q'
).interactive()
```

```python
from vega_datasets import data
population = data.population()
```

```python
alt.Chart(population).mark_bar().encode(
    x='year:O',
    y='sum(people):Q',
    color='sex:N'
)
```

```python
population['men_women'] = population['sex'].map({1: 'Men', 2: 'Women'})

alt.Chart(population).mark_bar().encode(
    x='year:O',
    y='sum(people):Q',
    color='men_women:N'
)
```
